### Transformation Text

```python
# Specify the folder path
output_folder_path = '../output/prose/Transformation.Text'
acc = []
count = 0

# Iterate over all files in the folder
for filename in os.listdir(output_folder_path):
    count += 1
    # Construct the full file path
    output_full_path = os.path.join(output_folder_path, filename)
    answer_path = "../data/foofah/Transformation.Text/"+filename
    # read in data
    with open(output_full_path, 'rb') as f:
        output_data = json.load(f)
    input_data, test_data = read_in_data(answer_path)
    output = output_data["result"]
    test = test_data[1][0]
    if output:
        acc_temp = 0
                # print(j,i)

        for k in range(min(len(output),len(test))):
            if output[k] == test[k]:                    
                acc_temp += 1
        acc.append(acc_temp/max(len(output),len(test)))
    else:
        acc.append(0)

    
```
